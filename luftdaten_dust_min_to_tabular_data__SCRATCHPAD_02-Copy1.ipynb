```python
#### testing interpolation/resampling, and what effect having NaN values has one the resampled vlues 
#### that is - if one has a row with NaN values, in the same time period as is being resampled, how will that affect the values? 
```

```python
# first, let's prepare a few lines of data … to resample
in_data_sample = in_data[:3]
in_data_sample
```

```python
in_data_sample.loc[ :, 'sensor_id'] = 200 
```

```python
in_data_sample
```

```python
# set the other values too 
in_data_sample.loc[ :, 'lat'] = 200 
in_data_sample.loc[ :, 'lon'] = 200 
in_data_sample.loc[ :, 'p1'] = 200 
in_data_sample.loc[ :, 'p2'] = 200 
```

```python
in_data_sample
```

```python
in_data_sample.loc[ 1:2, 'p1'] = np.NaN
in_data_sample.loc[ 1:2, 'p2'] = np.NaN
```

```python
in_data_sample
```

```python
# now let's try resampling 
in_data_sample__resampled = in_data_sample.resample( "5Min" ).mean()
in_data_sample__resampled
```

```python
# OK - this looks like the mean is not affected by the NAN values . 
# - LET'S TRY changing the p1 and p2 values a bit, to double-check that the averaging happens as I hope ( ie NaN values aren't included in the calculations )
```

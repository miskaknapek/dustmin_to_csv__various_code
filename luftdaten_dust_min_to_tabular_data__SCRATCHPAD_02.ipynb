```python
starttime = time.time()

found_minimal_value = 100000000000000000;

# for curr_sensor_id_index in range( unique_sensor_ids.shape[0] ) :
for curr_sensor_id_index in range( 2 ) :
        
        curr_sensor_id_and_row_num_i = unique_sensor_ids[ curr_sensor_id_index ]
        
        # -- fetch the time values 
        curr_sensor_data = in_data[ in_data['sensor_id'] == curr_sensor_id_and_row_num_i ]['timestamp']
        
        # -- find the minimal value in the array and 
        # (below) check if it's the absolutely lowest value
        # in the whole array
        curr_min_value = np.min( curr_sensor_data )
        
        # min check against the whole larger array
        if curr_min_value < found_minimal_value:
            found_minimal_value = curr_min_value
        
        # -- prepare for inserion - find the length of data
        curr_sensor_data_length = curr_sensor_data.shape[0]
        
        print("\n -- working on row "+str( curr_sensor_id_index )+"/"+str( unique_sensor_ids.shape[0] )+" sensor_id "+str( curr_sensor_id_and_row_num_i )+" - which has a length of "+str( curr_sensor_data_length ) )
        
        # -- insert the timestamp data into the larger timestapm numpy array 
        out_numpy_array_w_all_sensor_values[ curr_sensor_id_index ][ 0 : curr_sensor_data_length ] = curr_sensor_data
        
        # --- also inser save some sensor values 
        # fetch the sensor ( p1 ) values 
        curr_sensor_data_pt02 = in_data[  in_data['sensor_id'] == curr_sensor_id_and_row_num_i ]['p1']
        
        # insert the value into the larger sensor data array
        out_numpy_array_w_all_sensor_values_PART_TWO[ curr_sensor_id_index ][ 0 : curr_sensor_data_length ] = curr_sensor_data_pt02
        

print("\n ----- and that all took "+str( time.time() - starttime )+" seconds ")
print(" --------- and the found min value was "+str( found_minimal_value  ) ) 
    
```
